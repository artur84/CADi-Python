```python
t=np.array([10,10,2,3,9,9,1,1,5])
ind=np.argsort(t)
print(ind)
tsorted=t[ind]
print("tsorted: "+ str(tsorted))
tsorted[::-1]

np.where(tsorted > 2)
```

```python
# Visualización de matrices
import matplotlib.pyplot as plt

m=np.zeros((256,256))
for row in range(0, m.shape[0]):
    for col in range(0, m.shape[1]):
        m[row,col]=row + col

plt.figure(1)
plt.imshow(m)
plt.xlabel('col'); 
plt.ylabel('row')
plt.show()
```

```python
#Visualización de campos escalares
x = np.arange(-5, 5, 0.01)
y = np.arange(-5, 5, 0.01)
X, Y = np.meshgrid(x, y)
print("X: "+ str(X))
R = np.hypot(X, Y)
Z = np.sinc(R)
from matplotlib import cm
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
surf = ax.plot_surface(X, Y, Z, cmap=cm.seismic, linewidth=0, antialiased=0)
fig.colorbar(surf)
plt.show()
```
